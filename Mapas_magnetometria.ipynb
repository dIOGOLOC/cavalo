```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import shapefile as shp
from fatiando.vis import mpl
from fatiando import gravmag, gridder, utils
```

```python
#Importando a Tabela com os dados Magnetométricos:
tab=shp.Reader('/home/diogo/Ubuntu One/Material Mestrado ON/Artigos e Resumos Enviados/Resumo SBGf - 2013/Dados Magnetometricos/MAG_CLIP.shp')
```

```python
#Salvando a Tabela de dados como uma arquivo .txt para uma melhor manipulação:
with open('/home/diogo/Ubuntu One/Material Mestrado ON/Artigos e Resumos Enviados/Resumo SBGf - 2013/Dados Magnetometricos/tab.txt', 'w') as f:
    f.write('# ' + ' '.join(i[0] for i in tab.fields[2:-1]) + '\n')
    for l in tab.records():
        f.write(' '.join(str(i) for i in l[1:-1]))
        f.write('\n')
```

```python
#Carregando os arquivos .txt:
a=np.loadtxt('/home/diogo/Ubuntu One/Material Mestrado ON/Artigos e Resumos Enviados/Resumo SBGf - 2013/Dados Magnetometricos/tab.txt')

#Vizualizando os dados da tabela-tab.txt:
print(a[:,[0,1,13]])

#Inserindo valores a x,y e z:
x, y, z = a[:,[0,1,13]].T
```

```python
#Inserindo o valor do grid:
shape=(125, 125)

#Interpolando (interp=true) e plotando os valores do campo magnético total (z) diminuido do IGRF:
#Selecionando apenas a área do mapeamento geológico(set_area) e mantendo a escala com (axis('scaled')):
mpl.figure(figsize=(10,5))
mpl.axis('scaled')
mpl.contourf(x,y,z,shape, np.linspace(-70, 70, 100), interp=True, cmap=mpl.cm.jet)
mpl.plot(x,y,".k")
mpl.set_area([272000,283000,7703000,7709000])
mpl.colorbar(shrink=0.87)
mpl.m2km()

#Salvando a imagem:
mpl.savefig('/home/diogo/Ubuntu One/Material Mestrado ON/Artigos e Resumos Enviados/Resumo SBGf - 2013/Dados Magnetometricos/mapa_anomalia_magnetica.pdf')
```

```python
#Interpolando e plotando os valores do sinal analítico do campo magnético total diminuido do IGRF:
#Utiliza-se o numpy.ma.fix_invalid.ravel para preencher regiões sem valores com 0 para uma melhor vizualização:
#Usa-se o gridder.interp para interpolar o dado:

xg, yg, m = [np.ma.fix_invalid(i, fill_value=0).ravel() for i in gridder.interp(x, y, z, shape)]
print(m)
ansig = gravmag.fourier.ansig(xg, yg, m, shape)
print(ansig)

#Plotando os valores do sinal analítico (ansig) do campo magnético total diminuido do IGRF:
#Selecionando apenas a área do mapeamento geológico(set_area) e mantendo a escala com (axis('scaled')):

mpl.figure(figsize=(10,5))
mpl.axis('scaled')
mpl.contourf(xg, yg, ansig, shape, np.linspace(0, 0.8, 100))
mpl.colorbar(shrink=0.87)
mpl.set_area([276000,282350,7702968,7706402])
mpl.m2km()

#Salvando a imagem:
mpl.savefig('/home/diogo/Ubuntu One/Material Mestrado ON/Artigos e Resumos Enviados/Resumo SBGf - 2013/Dados Magnetometricos/mapa_sinal_analitico.png', dpi=600)
```
